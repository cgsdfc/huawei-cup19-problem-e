土壤含水量W，即各个深度的土壤湿度之和。

```python
df=pd.read_csv('./附件3-湿度-预处理.csv',index_col='time').drop(columns=[
    'month','year'
]).sort_index()
df=df.sum(axis=1)
df.name='土壤含水量（kg/m2）'
df.to_csv('土壤含水量W.csv')
W=df
df
```

```python
df=pd.read_csv('./附件8-气候-预处理.csv',index_col='time').drop(columns=[
    'month','year'
])['降水量(mm)'].sort_index()
df.to_csv('降水量P.csv')
P=df
df
```

```python
df=pd.read_csv('附件4-土壤蒸发量-预处理.csv',index_col='time').drop(columns=[
    'month','year'
])['土壤蒸发量(mm)'].sort_index()
df.to_csv('实际蒸发量Et_a.csv')
Eta=df
df
```

```python
df=pd.read_csv('附件6、植被指数-NDVI2012-2022年-预处理.csv').drop(columns=[
    '月份','年份'
]).rename(columns={'时间':'time'}).set_index('time')['植被指数(NDVI)'].sort_index()
df.to_csv('植被覆盖度cp.csv')
cp=df
df
```

```python
df=pd.read_csv('附件10-低层植被叶面积指数LAIL-预处理.csv'
              ).rename(columns={'日期':'time'}).set_index('time')[
    '低层植被(LAIL,m2/m2)'
].sort_index()
df.to_csv('叶面积指数LAI.csv')
LAI=df
df
```

```python
Vars=[W,P,Eta,cp,LAI]
df=pd.concat(Vars,axis=1).dropna()
```

```python
df
```

```python
df['土壤含水量差值（kg/m2）']=df['土壤含水量（kg/m2）'].diff()
df=df.dropna()
df
```

```python
df['累积降雨量（mm）']=df['降水量(mm)'].cumsum()
df
```

```python
df.columns
```

```python
W='土壤含水量（kg/m2）'
P='降水量(mm)'
Eta='土壤蒸发量(mm)'
cp='植被指数(NDVI)'
LAI='低层植被(LAIL,m2/m2)'
dW='土壤含水量差值（kg/m2）'
Rcum='累积降雨量（mm）'
```

```python
ICmax=0.935+0.498*df[LAI]-0.00575*df[LAI]**2
ICmax.name='植被的最大截流量（mm）'
ICmax_val=ICmax
ICmax=ICmax_val.name
```

```python
df[ICmax]=ICmax_val
```

```python
ICstore='植被截流量（mm）'
df[ICstore]=df[P]-df[Eta]-df[dW]
```

```python
df[ICstore][df[ICstore]<0]=0
```
